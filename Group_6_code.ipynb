```python
def diff(a,b):
    return a-b
df_1.insert(5,column='TravelTime',value=diff(df_1['Exit'],df_1['Entry']))
```

```python
def speed(a):
    return 223.2/a
df_1.insert(6,column='Speed(kmph)',value=speed(df_1['TravelTime']))
```

```python
new_col=['CumulativeEntryFlow']
df_1['CumulativeEntryFlow'] = 0
for i in range(87):
    for j in range(len(df_1)):
        if df_1['Entry'][j] <= df_1['Time Interval'][i]:
           df_1['CumulativeEntryFlow'][i] = df_1['CumulativeEntryFlow'][i]+1
df_1
```

```python
new_col=['EntryFlow/Interval']
df_1['EntryFlow/Interval'] = df_1['CumulativeEntryFlow'].shift(0).diff()
df_1['EntryFlow/Interval'][0] = 52
df_1
```
